```python
df3=df2.drop(['Sch_Require','Sch_Unit','Sch_Rub'],axis=1)
df3.head()
```

## Change DataTypes

```python
df3.dtypes
```

```python
df3['Sch_Date'] = pd.to_datetime(df3['Sch_Date']).dt.strftime('%d-%m-%Y')
```

```python
df3.head()
```

```python
int_list=['Sch_PlanQty','Sch_BatchNo','Sch_ActulQty','weight']
for i in int_list:
    df3[i]=df3[i].astype('int')
```

```python
df3['Time']=(df3['Sch_EndTime']-df3['Sch_StartTime'])
```

```python
df3.Time=df3['Time'].dt.total_seconds()/60
```

```python
df3.Time = df3.Time.astype('int')
```

```python
df4=df3.drop(['Sch_StartTime','Sch_EndTime'],axis=1)
df4.head()
```

## Adding 0 ahead of lotnumber

```python
df4['Sch_BatchNo']=df4['Sch_BatchNo'].astype('str')
```

```python
df4.reset_index(inplace=True,drop=True)
```

```python
df4['Sch_BatchNo'][2]
```

```python
df4.index
```

```python
for i in df4.index:
    if len(df4['Sch_BatchNo'][i])!=9:
        df4['Sch_BatchNo'][i]='0'+df4['Sch_BatchNo'][i]
```

## Replacing outliers

```python
df4['speed']=df4['Sch_ActulQty']/df4['Time']
```

```python
ol_index=df4[(df4['speed']>35) | (df4['speed']<5)].index
```

```python
df4.Time[25]
```

```python
for x in ol_index:
    df4.Time[x] = df4.Sch_ActulQty[x]/20
```

```python
df4.Time = df4.Time.astype('int')
```

```python
df5=df4.iloc[:,:-1]
df5
```

```python
df5.to_csv('total_production.csv',index=False)
```

# Production Achievement

```python
df6=df5.drop(['Sch_BatchNo','weight'],axis=1)
```

```python
df6['Plan_time']=df6['Sch_PlanQty']/20
df6['Plan_time']=df6['Plan_time'].astype('int')
df6.head()
```

```python
df7=df6.groupby(['Sch_Date','Sch_ItemCode'])['Plan_time','Time','Sch_PlanQty','Sch_ActulQty'].sum().reset_index()
```

```python
df7['Achievement_rate'] = (df7['Sch_ActulQty']/df7['Sch_PlanQty'])*100
df7['Achievement_rate'] = df7['Achievement_rate'].astype('int')
df7
```

```python
df7.to_csv('Achievement.csv',index=False)
```
