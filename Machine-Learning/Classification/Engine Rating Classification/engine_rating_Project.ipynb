# Null Value Handling

```python
df.isnull().sum()
```

```python
na_columns=[i for i in df.columns if df[i].isnull().sum()>0]
```

```python
for i  in na_columns:
    print(i,df[i].isnull().sum())
```

```python
un_column=[]
for i  in na_columns:
    if df[i].isnull().sum()>20000:
        un_column.append(i)
```

```python
len(un_column)
```

```python
un_column
```

```python
df1=df.drop(un_column,axis=1)
```

```python
df1.shape
```

```python
## null value.
x=[]
for i  in df1.columns:
    if df1[i].isnull().sum()>1:
        x.append(i)
```

```python
len(x)
```

```python
## check null values in categorial and numeric column.
cat=[]
num=[]
for i in x:
    if df1[i].dtype != 'o':
        cat.append(i)
    else:
        num.append(i)
```

```python
## value count
for i in cat:
    print(df1[i].mode())
```

```python
df2=df1.copy()
```

```python
df2
```

```python
### replace value 

for i in cat:
    df2[i]=df2[i].fillna("missing")
    df2[i+'_na'] =df2[i].apply(lambda x: 1 if x=="missing" else 0)
```

### Train Test Split

```python
x = df7.drop(['rating_engineTransmission'],axis=1)
y = df7['rating_engineTransmission']
```

```python
na_features = [i for i in df7.columns if i[-2:]=='na']
na_features
```

### Finding best features

```python
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2,f_regression
```

```python
fs= SelectKBest(chi2, k=10)
fit = fs.fit(xf,y)
```

```python
dfscores = pd.DataFrame(fit.scores_,columns=['score'])
dfcolumns = pd.DataFrame(xf.columns,columns=['specs'])
featureScores = pd.concat([dfcolumns,dfscores],axis=1)
print(featureScores.nlargest(10,'score')) 
```

```python
from sklearn.ensemble import ExtraTreesClassifier
model = ExtraTreesClassifier()
model.fit(xf,y)
```

```python
feat_importances = pd.DataFrame(model.feature_importances_, index=xf.columns,columns=['Weightage'])
impt_df = feat_importances.nlargest(15,'Weightage').reset_index()
```

```python
a = list(impt_df['index'])
a
```

```python
impt_df.plot(kind='bar')
```

```python
df8 = df7[a]
x = df8.drop(['rating_engineTransmission'],axis=1)
y = df8['rating_engineTransmission']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.25)
```

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier()
rf.fit(x_train,y_train)
```

```python
rf.score(x_test,y_test)
```
