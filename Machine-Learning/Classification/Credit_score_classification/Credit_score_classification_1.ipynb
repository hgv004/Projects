```python
ol_col=[i for i in df18.columns if len(df17[i].unique())>5]
ol_col
```

```python
okl=[]
oll=[]
for i in ol_col:
  ll,ul=np.percentile(df18[i],(10,90))
  iqr=ul-ll
  lr=ll-(1.5*iqr)
  ur=ul+(1.5*iqr)
  ol_ind=df18[((df18[i]>ur) | (df18[i]<lr))].index
  oll.extend(ol_ind)
print('According to IQR method there are total:', len(oll),'outliers')
```

```python
ol_set=set(oll)
ol_list=list(ol_set)
print(len(ol_list))
```
